```python
iris_data = pd.read_csv('./data/iris.csv', index_col='Id')
iris_data.head()
```

```python
# Column 데이터를 불러오는 방법

#그냥 불러오기
Species = iris_data.Species

# ['cloumn name']
Species_bracket = iris_data['Species']

print(Species)
print('')
print(Species_bracket)
```

```python
#  Row 데이터를 불러오는 방법 

#loc = Index 이름으로 불러오기
Species_loc = iris_data.loc[1] # 1번 id가 있는 row 불러오기

#iloc = Index 위치로 불러오기  / i = integer
Species_iloc = iris_data.iloc[0] # 1번 id가 있는 row는 0번임.


print(Species_loc)
print('')
print(Species_iloc)
```

```python
# 특정 행렬 값 찾기

# row => row index
iris_data.iloc[0].loc['Species']

# column => row 위치
iris_data.Species.iloc[0]

# column => low 이름
iris_data['Species'].loc[1]

# at 매소드[row,columns]
iris_data.at[1,'Species']
```

```python
#slicing
my_arr = np.array([[2,3,4,5],[10,11,12,13],[23,24,25,26],[41,42,43,44]])

# [row, column]
my_arr[0,0]
my_arr[0,:]
my_arr[:,0]
my_arr[:,:]

#[row, column 범위]
my_arr[0:4,0:2]
```

```python
#pandas Slicing
iris_data.loc[20:80,'PetalLengthCm':'PetalWidthCm']
iris_data.iloc[19:80,2:4]
```

```python
a = iris_data.sort_values('PetalLengthCm')
a
```

```python
# Reset index를 쓰면 index가 새로 생김
b = iris_data.sort_values('PetalLengthCm').reset_index()
b
```

```python
#filtering 방법. Boolean masking
k = b.Species == 'Iris-setosa'
k
```

```python

# Filter 설정한 뒤 특정 column 추출
b[k].PetalWidthCm
```

```python
print('PetalwithCm Mean : {}'.format(np.mean(b[k].PetalWidthCm)))
print('PetalLengthCm Mean : {}'.format(round(np.mean(b[k].PetalLengthCm),2)))
```
