```python
import pandas as pd
import numpy as np
```

```python
football_events = pd.read_csv('./data/events.csv', index_col = 'sort_order')
football_events.describe()
```

```python
football_events.head(5)
```

```python
team_1 = (football_events.event_team == 'Tottenham')
team_2 = (football_events.event_team == 'Arsenal')
team_3 = (football_events.event_team == 'Paris Saint-Germain')
football_column_drop = football_events.drop(['id_odsp','event_type','event_type2','side','location','text'], axis=1)
football_filter = football_column_drop.loc[team_1 | team_2 | team_3, : ]
football_filter.set_index('id_event') # changes a column into index
football_filter.describe() 
```

```python
# chnges the name of some values.
def change_name(x) :
    if x == 'heungmin son' or x =='son heungmin' :
        x = x.replace('heungmin son', 'Son Heung-min' )
        x = x.replace('son heungmin', 'Son Heung-min' )
    return x
football_filter.player = football_filter.player.apply(lambda x: change_name(x))
```

```python
a = football_filter.player.unique()
b = [print(player) for player in a if 'Son' in str(player)]
```
