```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
x = np.arange(1,101, 10)
y = 4 *x 
plt.plot(y,label='linear')


plt.subplot(1,2,2)
x = np.arange(1,101, 10) **2
y = 4 *x 
plt.plot(y, color='tab:red',label='polynomial')


# plt.legend(bbox_to_anchor=(1.05,1.05))
```
