```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# 한글 폰트 설정
plt.rc("font", family = "Malgun Gothic")
sns.set(font="Malgun Gothic", 
rc={"axes.unicode_minus":False}, style='darkgrid')
```

```python
# %memit
academy_df = pd.read_csv('C:/Users/679oo/Downloads/academy.csv',encoding='cp949', dtype=str)
academy_df.info()
BM = academy_df['학교'] != '육군 3사관학교'
academy_df = academy_df[BM]
```

```python
# BM = academy_df['남자 경쟁률'] == '-'
academy_df['남자 경쟁률'] = academy_df['남자 경쟁률'].mask(academy_df['남자 경쟁률'] == '-', '0')
```

```python
academy_df['전체 경쟁률'] = np.array([ re.sub(":1",'',i)for i in academy_df['전체 경쟁률']]).astype(float)
academy_df['남자 경쟁률'] = np.array([ re.sub(":1",'',i)for i in academy_df['남자 경쟁률']]).astype(float)
academy_df['여자 경쟁률'] = np.array([ re.sub(":1",'',i)for i in academy_df['여자 경쟁률']]).astype(float)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='년도',y='여자 경쟁률', hue='학교', data=academy_df)
```
